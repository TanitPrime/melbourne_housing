```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
# Importing file and previewing data
melbourne_housing= pd.read_csv("melb_data.csv")
print(melbourne_housing.describe())
print(melbourne_housing.head())
```

```python
# Identifying numeric and categorical data columns
ncols= [col for col in melbourne_housing.columns if melbourne_housing[col].dtype != "object" ]
catcols= melbourne_housing.columns[~melbourne_housing.columns.isin(ncols)]

# Extracting numeric data for correlation matrix
melbourne_housing_numeric= melbourne_housing[ncols]
```

```python
sns.heatmap(melbourne_housing_numeric.corr(), annot= True)
plt.show()
```

```python
# Let's try and visualize the relationship of price vs Rooms, Bathroom and YearBuilt as they seem to have a relatively high explanatory value

g= plt.figure(figsize=(2,5))
g= sns.FacetGrid(melbourne_housing, col= "Price")
g.map(sns.regplot, "Rooms", "Bathroom")
g.fig.suptitle("Price vs Rooms and Bathroom")
```
